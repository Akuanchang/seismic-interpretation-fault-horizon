```python
# Necessary imports
import sys
import numpy as np

sys.path.append('../..')
from seismiqb import plot_image
```

```python
x = np.linspace(1, 100, 1000)

oscillation = lambda x: np.sin(x) / x + np.random.normal(scale=0.1, size=len(x))
rand = lambda : np.random.rand()

y0 = -np.log(x) * rand() + oscillation(x)
y1 = -np.log(x) * rand() + oscillation(x)
```

```python
figure_params = {
    'mode': 'curve',
    'figsize': (20, 7)
}

annotation_params = {
    'title_label': 'Simulated loss'
}

plot_image(y0, **figure_params, **annotation_params)
```

```python
curve_params = {'average': True}

plot_image(y0, **figure_params, **curve_params, **annotation_params)
```

```python
plot_image([y0, y1], **figure_params, **curve_params, **annotation_params)
```

```python
plot_image([y0, y1], separate=True, **figure_params, **curve_params, **annotation_params)
```

```python
figure_params['figsize'] = (25, 7)

color_0, color_1 = 'skyblue', 'sandybrown'
curve_params['color'] = [color_0, color_1, [color_0, color_1]]

plot_image([y0, y1, [y0, y1]], **figure_params, **curve_params, **annotation_params)
```
