```python
# Necessary imports
import sys
import numpy as np

sys.path.append('../..')
from seismiqb import plot_image
```

```python
shape = (50, 50)
rand = lambda : np.random.rand()
image_0 = np.random.normal(loc=rand(), scale=rand(), size=shape)
image_1 = np.random.laplace(loc=rand(), scale=rand(), size=shape)
```

```python
plot_image([image_0, image_1], figsize=(16,8), cmap='Greys', colorbar=True, separate=True)
```

```python
figure_params = {
    'mode' : 'hist',
    'figsize' : (15,5)
}

annotation_params = {
    'title_label' : 'Image histogram',
    'xlabel': 'Domain',
    'ylabel': 'Counts'
}

plot_image([image_0, image_1], separate=True, **annotation_params, **figure_params)
```

```python
figure_params['figsize'] = (8, 5)

plot_image([image_0, image_1], **annotation_params, **figure_params)
```

```python
hist_params = {
    'alpha': [1., .8]
}

plot_image([image_0, image_1], **hist_params, **annotation_params, **figure_params)
```

```python
annotation_params['legend_label'] = ['image_0', 'image_1']

plot_image([image_0, image_1], **hist_params, **annotation_params, **figure_params)
```

```python
hist_params['cumulative'] = True

plot_image([image_0, image_1], **hist_params, **annotation_params, **figure_params)
```

```python
hist_params['color'] = ['tan', 'thistle']
hist_params['cumulative'] = False
hist_params['histtype'] = 'step'

plot_image([image_0, image_1], **hist_params, **annotation_params, **figure_params)
```

```python
figure_params['figsize'] = (25, 5)
color_0, color_1 = 'firebrick', 'forestgreen'
alpha_0, alpha_1 = 1., .7
hist_params['color'] = [[color_0], [color_0, color_1], [color_1]]
hist_params['alpha'] = [[alpha_0], [alpha_0, alpha_1], [alpha_0]]
hist_params['histtype'] = 'bar'

annotation_params['legend_label'] = [[None], ['image_0', 'image_1'], [None]]

plot_image([image_0, [image_0, image_1], image_1], **hist_params, **annotation_params, **figure_params)
```
