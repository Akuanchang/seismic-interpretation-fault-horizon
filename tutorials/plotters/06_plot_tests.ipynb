```python
import sys
import numpy as np

PROJECT_PATH = '../' * 3
sys.path.insert(0, f'{PROJECT_PATH}seismiqb')

from seismiqb import plot_image
```

```python
arr = np.random.rand(10, 10)
```

```python
data = arr
title = 't'
cmap = 'ocean'
plot_image(data=data, title=title, cmap=cmap)
```

```python
data = [arr]
title = ['t']
cmap = ['ocean']
plot_image(data=data, title=title, cmap=cmap)
```

```python
data = [arr, arr]
title = ['t', 'y']
cmap = ['ocean', 'viridis']
plot_image(data=data, title=title, cmap=cmap)
```

```python
data = [arr, arr]
title = ['t', 'y']
cmap = ['ocean', 'viridis']
plot_image(data=data, title=title, cmap=cmap, separate=True)
```

```python
data = [arr, [arr, arr]]
title = ['t', 'y']
cmap = ['ocean', 'viridis']
plot_image(data=data, title=title, cmap=cmap)
```

```python
data = [arr, [arr, arr]]
title = ['t', ['y', 'x']]
cmap = ['ocean', ['viridis', 'Reds']]
plot_image(data=data, title=title, cmap=cmap)
```

```python
data = [arr, [arr, arr > .5]]
title = ['t0', 't1']
cmap = ['ocean', ['viridis', 'firebrick']]
plot_image(data=data, title=title, cmap=cmap)
```
