```python
import matplotlib.pyplot as plt
```

```python
ncols = np.random.randint(1, 4)
nrows = np.random.randint(1, 4)
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(10 * nrows, 10 * ncols))
axes = axes.flatten() if isinstance(axes, np.ndarray) else [axes]
shape = tuple(np.random.randint(2, 10, 2))
arr = np.ones(shape)
for i in range(ncols*nrows):
    axes[i].imshow(np.ones(np.random.randint(2, 10, 2)))
    axes[i].set_title('title title title title title title title title')
```

```python
renderer = fig.canvas.get_renderer()
transformer = fig.dpi_scale_trans.inverted()
get_inner_bbox = lambda obj: obj.get_window_extent(renderer=renderer).transformed(transformer)
get_outer_bbox = lambda obj: obj.get_tightbbox(renderer).transformed(transformer)
```

```python
boxes = [get_outer_bbox(ax) for ax in axes]
```

```python
heights = [b.height for b in boxes]
widths = [b.width for b in boxes]
```

```python
heights_reshaped = np.array(heights).reshape(nrows, ncols)
widths_reshaped = np.array(widths).reshape(nrows, ncols)
```

```python
height = heights_reshaped.max(axis=1).sum()
width = widths_reshaped.max(axis=0).sum()

ratio = height / width

width = 8 * ncols
height = width * ratio
width, height
```

```python
fig.set_size_inches((width, height))
fig.tight_layout()
```
